```python
# sketch of a test code. scores each word, higher score is higher difficulty

def getSyllable(word): # checks for syllables in word -- very basic, can look up more complex code
    word = word.lower()
    count = 0
    vowels = "aeiouy"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
    if word.endswith("e"):
        count -= 1
    if count == 0:
        count += 1
    return count
index = -1
score = [] # creating a list to add to the table created with pandas
for i in words['word']:
    index += 1
    diff_score = 0
    if len(i) > 6: # checks for length of word
        diff_score += 1
    if words.iloc[index, 1] < 40.0: # checks for success rate from sample
        diff_score += 1
    if words.iloc[index, 12] < 30.0: # checks for probability of successfully translating word
        diff_score += 1
    if words.iloc[index, 13] != i: # checks if word is in root form or has suffix
        diff_score += 1
    if getSyllable(i) >= 3: # checks a rough estimate of syllables in word
        diff_score += 1
    # need a test here that runs word through a series of reading sources with a low Flesch-Kincaid score -- harder words most likely will not be contained and can gain a point 
    score.append(diff_score) # adding the words score to the list
words['score'] = score # adding final list to the table
print(words['score'])
```
